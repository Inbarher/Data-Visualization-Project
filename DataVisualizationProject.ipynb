נתחיל מהמרת עמודות הזיכרון והאחסון לנומריות

```python
def to_gb(val):
    if pd.isnull(val):
        return None
    val = str(val).strip()
    # חפש מספר לפני GB או MB
    import re
    match = re.search(r'(\d+(\.\d+)?)\s*(GB|MB)', val)
    if match:
        num = float(match.group(1))
        unit = match.group(3)
        if unit == 'GB':
            return num
        elif unit == 'MB':
            return num / 1024
    # אם אין התאמה, החזר None
    return None
```

נשנה את העמודות הקיימות

```python
df['Memory'] = df['Memory'].apply(to_gb)
df['Storage'] = df['Storage'].apply(to_gb)
```

לא טוב כיוון שנוצרו ערכים שאינם קיימים במציאות. אין פאלפון עם 1.5 זיכרון. לכן כשיש מספר אי זוגי נבחר באקראי אחד מהמספרים האמצעיים מהם ולא נעשה ממוצע בין  השניים האמצעיים כמו שעושים בחישוב חציון

```python
import numpy as np

def fill_with_random_median(series):
    non_na = series.dropna().sort_values().values
    n = len(non_na)
    if n == 0:
        return series  # אם אין בכלל ערכים – לא נוגעים
    if n % 2 == 1:
        median_val = non_na[n // 2]  # אם מספר אי-זוגי – פשוט החציון
    else:
        # אם מספר זוגי – בחר באקראי אחד מבין שני האמצעיים
        mid1, mid2 = non_na[n // 2 - 1], non_na[n // 2]
        median_val = np.random.choice([mid1, mid2])
    return series.fillna(median_val)
```

בנוסף על מנת לדייק נעשה זאת לפי החציון של אותו מותג טלפון ולא של כלל הדאטה

```python
# df['Memory'] = df.groupby('Brands')['Memory'].transform(fill_with_random_median)
# df['Storage'] = df.groupby('Brands')['Storage'].transform(fill_with_random_median)
```

```python
df['Memory'] = fill_with_random_median(df['Memory'])
df['Storage'] = fill_with_random_median(df['Storage'])
```

```python
def simplify_color(color):
    color = str(color).strip().lower()
    if 'black' in color:
        return 'Black'
    elif 'blue' in color:
        return 'Blue'
    elif 'white' in color:
        return 'White'
    elif 'red' in color:
        return 'Red'
    elif 'gold' in color:
        return 'Gold'
    elif 'green' in color:
        return 'Green'
    elif 'grey' in color or 'gray' in color:
        return 'Gray'
    elif 'silver' in color:
        return 'Silver'
    elif 'pink' in color:
        return 'Pink'
    elif 'purple' in color or 'violet' in color or 'lilac' in color:
        return 'Purple'
    elif 'brown' in color:
        return 'Brown'
    elif 'orange' in color:
        return 'Orange'
    elif 'yellow' in color:
        return 'Yellow'
    else:
        return 'Other'

# עדכון העמודה הקיימת
df['Colors'] = df['Colors'].apply(simplify_color)
```

KMeans הפעלת

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

wcss = []
silhouette_scores = []
K_range = range(2, 11)  # נבדוק בין 2 ל-10 קלאסטרים

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_kmeans)
    wcss.append(kmeans.inertia_)  # WCSS
    score = silhouette_score(X_kmeans, labels)
    silhouette_scores.append(score)

# גרף Elbow
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.plot(K_range, wcss, 'bo-', markersize=8)
plt.xlabel('Number of clusters (k)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method For Optimal k')

# גרף Silhouette Score
plt.subplot(1, 2, 2)
plt.plot(K_range, silhouette_scores, 'go-', markersize=8)
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score For Optimal k')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# --- 1. קריאת הנתונים (החלף לדרך שבה אתה טוען את הנתונים שלך) ---


# --- 2. בחירת עמודות רלוונטיות ---
features = [
    'Brands', 'Models', 'Colors', 'Memory', 'Storage', 'Rating', 'Is_Pro_Model', 'Is_Plus_Model', 'Is_Max_Model', 'Is_Mini_Model'
]
target = 'Selling Price'
cluster_col = 'Cluster'

# --- 3. פיצול ל-train/test ---
X = df[features].copy()
y = df[target].copy()
cluster = df[cluster_col].copy()

X_train, X_test, y_train, y_test, cluster_train, cluster_test = train_test_split(
    X, y, cluster, test_size=0.2, random_state=42
)

# --- 4. קידוד עמודות קטגוריות ובינאריות ---
categorical_cols = ['Brands', 'Models', 'Colors']
numeric_cols = [col for col in features if col not in categorical_cols]

# One-Hot Encoding
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoder.fit(X_train[categorical_cols])

X_train_cat = encoder.transform(X_train[categorical_cols])
X_test_cat = encoder.transform(X_test[categorical_cols])

# איחוד עם עמודות נומריות
X_train_num = X_train[numeric_cols].values
X_test_num = X_test[numeric_cols].values

X_train_all = np.hstack([X_train_num, X_train_cat])
X_test_all = np.hstack([X_test_num, X_test_cat])

# --- 5. PCA ---
n_components = 5  # שנה לפי הצורך
pca = PCA(n_components=n_components, random_state=42)
X_train_pca = pca.fit_transform(X_train_all)
X_test_pca = pca.transform(X_test_all)

# --- 6. הוספת עמודת ה-Cluster ---
X_train_pca_cluster = np.hstack([X_train_pca, cluster_train.values.reshape(-1, 1)])
X_test_pca_cluster = np.hstack([X_test_pca, cluster_test.values.reshape(-1, 1)])

# --- 7. Random Forest ---
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train_pca_cluster, y_train)
y_pred_rf_pca = rf.predict(X_test_pca_cluster)

# --- 8. מדדים והדפסה ---
mae_rf_pca = mean_absolute_error(y_test, y_pred_rf_pca)
mse_rf_pca = mean_squared_error(y_test, y_pred_rf_pca)
rmse_rf_pca = np.sqrt(mse_rf_pca)
r2_rf_pca = r2_score(y_test, y_pred_rf_pca)

print(f"Random Forest with PCA + KMeans Cluster Results:")
print(f"MAE: {mae_rf_pca:.2f}")
print(f"MSE: {mse_rf_pca:.2f}")
print(f"RMSE: {rmse_rf_pca:.2f}")
print(f"R2 Score: {r2_rf_pca:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. קידוד חם לעמודות קטגוריאליות: Brands, Colors, Models
df_encoded = pd.get_dummies(df, columns=['Brands', 'Colors', 'Models'], drop_first=True)

# 2. בחירת פיצ'רים
features = ['Memory', 'Storage', 'Rating', 'Is_Pro_Model', 'Is_Plus_Model', 'Is_Max_Model', 'Is_Mini_Model']

# הוספת כל העמודות שנוצרו מקידוד חם (התחלת השמות ב-Brands_, Colors_, Models_)
features += [col for col in df_encoded.columns if col.startswith('Brands_') or col.startswith('Colors_') or col.startswith('Models_')]

X = df_encoded[features]
y = df_encoded['Selling Price']  # או השם המדויק של עמודת מחיר המכירה שלך

# 3. סקלינג
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"מספר מימדים לפני PCA: {X_scaled.shape[1]}")

# 4. PCA (שומר על 95% שונות)
pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print(f"מספר מימדים אחרי PCA: {X_pca.shape[1]}")
print(f"שונות מצטברת שנשמרה: {np.sum(pca.explained_variance_ratio_):.2%}")

# 5. KMeans - אשכולות
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_pca)
df_encoded['Cluster'] = clusters

# הוספת האשכול כמאפיין נוסף
X_final = np.hstack([X_pca, clusters.reshape(-1, 1)])

# 6. חלוקה לטריינינג וטסט
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)

# 7. Random Forest Regression לחיזוי מחיר המכירה
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

# 8. הדפסת מדדים
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("📊 Random Forest Regression Results:")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R2 Score: {r2:.4f}")
```
